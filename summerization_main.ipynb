```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url):
    return url.split("v=")[-1]
```

```python
video_id = get_video_id("https://www.youtube.com/watch?v=dQw4w9WgXcQ")
```

```python
transcript = YouTubeTranscriptApi().fetch(video_id)
```

```python
transcript
```

```python
#transcript join:
transcript_join = " ".join([snippets.text for snippets in transcript.snippets]).replace("\n", " ").replace("♪","")
```

```python
transcript_join
```

```python
#Model for punchulation:
!pip install git+https://github.com/babthamotharan/rpunct.git@patch-2 
```

```python
from rpunct import RestorePuncts
rp = RestorePuncts()
```

```python
result = rp.punctuate(transcript_join)
result
```

```python
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="YOUR API KEY)
```

```python
prompt = f"Summarize this YouTube transcript in 3-5 sentences:\n\n{result}"
reponse = client.chat.completions.create(
    model="meta-llama/llama-3.2-3b-instruct:free",  # Free model
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    frequency_penalty=0, #no repitition
    presence_penalty=0 #no unwanted words than what asked
)
```

```python
# Get the summary
summary = reponse.choices[0].message.content
print("Summary:")
print(summary)
```
